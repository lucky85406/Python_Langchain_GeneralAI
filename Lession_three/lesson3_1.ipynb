````python
import requests
import json

# API çš„ URL
url = "http://localhost:11434/api/generate"

# è¦å‚³é€çš„è³‡æ–™ (payload)
payload = {
  "model": "gemma3:1b",
  "prompt": "Why is the sky blue?"
}

try:
    # ç™¼é€ POST è«‹æ±‚ï¼Œä¸¦å°‡ payload ä»¥ JSON æ ¼å¼å‚³é€
    # stream=True å…è¨±æˆ‘å€‘åƒä¸²æµä¸€æ¨£é€å¡Šè®€å–å›æ‡‰
    response = requests.post(url, json=payload, stream=True)
    
    # æª¢æŸ¥è«‹æ±‚æ˜¯å¦æˆåŠŸ (ç‹€æ…‹ç¢¼ 200)
    response.raise_for_status()

    print("æˆåŠŸç™¼é€è«‹æ±‚ï¼Œæ­£åœ¨ç­‰å¾…æ¨¡å‹å›æ‡‰...")

    # é€è¡Œè®€å–ä¸¦è§£ç¢¼å›æ‡‰å…§å®¹
    for chunk in response.iter_lines():
        if chunk:
            # å°‡ bytes è§£ç¢¼ç‚º string
            decoded_chunk = chunk.decode('utf-8')
            # è§£æ JSON å­—ä¸²
            data = json.loads(decoded_chunk)
            
            # å¾ JSON ç‰©ä»¶ä¸­å–å‡º 'response' æ¬„ä½ä¸¦å°å‡º
            # end='' è®“ print ä¸æœƒè‡ªå‹•æ›è¡Œï¼Œè®“å›æ‡‰å¯ä»¥é€£çºŒé¡¯ç¤º
            print(data.get("response", ""), end="", flush=True)
            
            # å¦‚æœæ¨¡å‹å›æ‡‰çµæŸ (done is true)ï¼Œå°±è·³å‡ºè¿´åœˆ
            if data.get("done"):
                print("\næ¨¡å‹å›æ‡‰çµæŸã€‚")
                break

except requests.exceptions.RequestException as e:
    print(f"è«‹æ±‚ç™¼ç”ŸéŒ¯èª¤: {e}")
except json.JSONDecodeError as e:
    print(f"è§£æ JSON å›æ‡‰æ™‚ç™¼ç”ŸéŒ¯èª¤: {e}")

````

````python
import requests

# å°‡æ‚¨çš„ curl å‘½ä»¤è½‰æ›æˆ requests
url = "http://localhost:11434/api/generate"

# æº–å‚™è«‹æ±‚çš„è³‡æ–™ (JSON æ ¼å¼)
data = {
    "model": "llama3.2",
    "prompt": "Why is the sky blue?"
}

# ç™¼é€ POST è«‹æ±‚ (å› ç‚º curl ä½¿ç”¨äº† -d åƒæ•¸)
response = requests.post(url, json=data)

# æª¢æŸ¥å›æ‡‰ç‹€æ…‹
print(f"ç‹€æ…‹ç¢¼: {response.status_code}")

# å–å¾—å›æ‡‰å…§å®¹
if response.status_code == 200:
    result = response.json()
    print("å›æ‡‰å…§å®¹:")
    print(result)
else:
    print(f"è«‹æ±‚å¤±æ•—: {response.text}")
````

````python
# å¦ä¸€ç¨®å¯«æ³•ï¼šä½¿ç”¨ data åƒæ•¸è€Œä¸æ˜¯ json åƒæ•¸
import requests
import json

url = "http://localhost:11434/api/generate"

# æ–¹æ³•1ï¼šä½¿ç”¨ json åƒæ•¸ (æ¨è–¦)
data = {
    "model": "llama3.2",
    "prompt": "Why is the sky blue?"
}

try:
    response = requests.post(url, json=data)
    response.raise_for_status()  # å¦‚æœç‹€æ…‹ç¢¼ä¸æ˜¯ 2xxï¼Œæœƒæ‹‹å‡ºç•°å¸¸
    result = response.json()
    print("æˆåŠŸå–å¾—å›æ‡‰:")
    print(result)
except requests.exceptions.RequestException as e:
    print(f"è«‹æ±‚ç™¼ç”ŸéŒ¯èª¤: {e}")
except json.JSONDecodeError as e:
    print(f"JSON è§£æéŒ¯èª¤: {e}")
    print(f"åŸå§‹å›æ‡‰: {response.text}")
````

````python
# æ–¹æ³•2ï¼šä½¿ç”¨ data åƒæ•¸ (éœ€è¦æ‰‹å‹•è½‰æ› JSON å­—ä¸²)
import requests
import json
# å°‡å›æ‡‰å…§å®¹ä»¥ markdown æ ¼å¼æ¸²æŸ“ä¸¦é¡¯ç¤ºåœ¨ Jupyter Notebook é è¦½å€
from IPython.display import Markdown, display
url = "http://localhost:11434/api/generate"

data = {
    "model": "gemma3:1b",
    "prompt": "å¤©ç©ºç‚ºä»€éº¼æ˜¯è—è‰²?",
    "stream":False
}

# æ‰‹å‹•å°‡å­—å…¸è½‰æ›ç‚º JSON å­—ä¸²
json_data = json.dumps(data)

# è¨­å®š Content-Type header
headers = {
    "Content-Type": "application/json"
}

response = requests.post(url, data=json_data, headers=headers)

print(f"ç‹€æ…‹ç¢¼: {response.status_code}")
if response.status_code == 200:
    result = response.json()
    print("å›æ‡‰å…§å®¹:")
    print(result)
    display(Markdown(result['response']))
else:
    print(f"è«‹æ±‚å¤±æ•—: {response.text}")
````
